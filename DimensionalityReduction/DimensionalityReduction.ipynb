## Introduction

Unsupervised machine learning and visualization help to assist you with data familiarity -- getting to know your data-- as well as assisting in the determination of patterns or relationships among unlabeled data samples. A data set with two dimensions is fairly simple to plot. Even data in three dimensions can be simple to plot, but what if you are working with a data set that has four or five dimensions? In this case, we use dimensionality reduction.

Dimensionality Reduction is a technique used in machine learning that takes a large, multi-dimensional data set and reduces the data into a lower amount dimensions (i.e. two or three). This allows us to see patterns or relationships in the data that you could not have seen in its four or more dimensional form because the reduced representation still manages to retain some meaningful properties that the original data contained. For example, if the data seems to be clustered, this would indicate a similarity or similarities in the features which hints towards classifications in the data set. 

![Nonlinear Dimensionality Reduction](NLDR.png)

This technique is helpful under many circumstances, such as the reduction of time for training estimators, or the reduction of the difficulties associated with having to imagine a more than three dimensions. This is also known as the *curse of dimensionality*.

```python
# Import the packages
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
%matplotlib inline
```

### Import the Data

With our packages imported, we are now able to import our data set

```python
# Import data
digits_data = load_digits()
digits_data.data.shape
```

```python
for i in range(0,3):
    plt.gray()
    plt.imshow(digits_data.images[i])
    plt.show()
```

---
### Reduce to `n_components = 2`

Now we will import our estimator. We will be using the t-distributed stochastic neighbor embedding (TSNE) estimator. This estimator is based on the Stochastic Neighbor Embedding, and is a nonlinear dimensionality reduction technique that is good for embedding large-dimensional data. It works in two stages:

1. TSNE constructs a probability distribution over pairs of large-dimensional objects in a way that similar objects are given a higher probability and those without similarities are assigned a lower probability.


2. TSNE defines a similar probability distribution over the points in the lower dimensions and minimizes the [Kullback-Leibler divergence](https://en.wikipedia.org/wiki/Kullback%E2%80%93Leibler_divergence) (KL divergence) between the two distributions with respect to the locations of the points in the dimensional map.

Once our estimator is imported, we can fit our data and display the size to show the dimension reduction, as seen in the code cell below.

```python
# Import estimator
tsne = TSNE(n_components=2, random_state=9)

reduced_digits_to2 = tsne.fit_transform(digits_data.data)
print("The shape of the reduced data set is ", reduced_digits_to2.shape)
```

---
#### Visualize the Reduced Data

With our data now reduced to two dimensions, we can use a scatter plot to display the data.

```python
redDig_to2_plot = plt.scatter(reduced_digits_to2[:,0], 
                          reduced_digits_to2[:, 1], c=digits_data.target, 
                         cmap=plt.cm.get_cmap('viridis', 10))
```

---
### Reduce to `n_components = 3`

```python
# Import estimator
tsne = TSNE(n_components=3, random_state=11)

reduced_digits_to3 = tsne.fit_transform(digits_data.data)
print("The shape of the reduced data set is ", reduced_digits_to3.shape)
```

---
#### Visualization

With our data once again reduced, except this time to three dimensions, we can use a scatter plot to display the data again.

```python
fig = plt.figure(figsize=(10, 15))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(reduced_digits_to3[:,0], reduced_digits_to3[:, 1], reduced_digits_to3[:, 2], 
           c=digits_data.target , cmap=plt.cm.get_cmap('viridis', 10))
plt.show()
```
